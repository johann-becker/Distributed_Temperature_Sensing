```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import datetime
from datetime import timedelta
```

```python
# User Inputs:
fname='DEC2_house_full.csv'
# Position of nodes along the fiber
Nodes=[4.845, 9.675, 14.505, 20.86, 26.198, 30.774, 37.383, 44.5, 54.668, 63.565, 72.462, 82.376]
# Name of Nodes (for plotting)
Node_names=['Driveway','Gravel','Pavement1','Grass','Pavement2','Leaves','Path (Top)', 'Path (Bottom)', 'Pool', 'Pool Weights','Patio', 'backgrass']
```

```python
#Load and format data
fiber= pd.read_csv(fname,header=0)
fiber.set_index(fiber['Distance'],inplace=True)
fiber=fiber.drop(columns=['Distance'])
fiber=fiber.transpose()
```

```python
#Load and format time series
time= pd.read_csv('DEC2_house_full.csv',header=None)

time = time.loc[0,1:]
time = pd.to_datetime(time)
time = time - timedelta(hours=5) #Corrects for time zone conversion
```

```python
# Plot multiple/all Nodes of interest on same figure
plt.figure(figsize=(15,8))
#plt.plot(time, fiber[Nodes[0]], label=Node_names[0])
plt.plot(time, fiber[Nodes[1]],label=Node_names[1])
plt.plot(time, fiber[Nodes[2]],label=Node_names[2])
plt.plot(time, fiber[Nodes[3]],label=Node_names[3])
#plt.plot(time, fiber[Nodes[4]],label=Node_names[4])
plt.plot(time, fiber[Nodes[5]],label=Node_names[5])
#plt.plot(time, fiber[Nodes[6]],label=Node_names[6])
plt.plot(time, fiber[Nodes[7]],label=Node_names[7])
plt.plot(time, fiber[Nodes[8]],label=Node_names[8])
#plt.plot(time, fiber[Nodes[9]],label=Node_names[9])
#plt.plot(time, fiber[Nodes[6]],label=Node_names[10])
#plt.plot(time, fiber[Nodes[7]],label=Node_names[11])
plt.xticks(rotation=45)
plt.legend()
plt.title('Ground Cover Temperture Variation')
plt.xlabel('Date & Time')
plt.ylabel('Temperature $^\circ$C');
plt.savefig('DEC2_house_nodes.png')
```
