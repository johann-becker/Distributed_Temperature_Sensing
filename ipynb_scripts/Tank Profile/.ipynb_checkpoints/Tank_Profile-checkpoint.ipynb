```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import datetime
```

```python
#User Inputs:
fname = 'OCE_AT.csv'
length = [83.139, 86.952] #Length measurements of fiber coresponding to subject area -- [ start, end ]
outfname = 'OCE_AT.png' #Output file name, including file type
```

```python
#Load and format data
profile= pd.read_csv(fname,header=1)
profile.set_index(profile['distance'],inplace=True)
```

```python
#Plot the temperature differences along the fiber
plt.figure(figsize=(5,10))
plt.plot(profile['temperature'][length[0]:length[1]],profile['distance'][length[0]:length[1]])
plt.gca().invert_yaxis()
plt.xlabel('Temperature Difference $^\circ$C')
plt.ylabel('Distance from Fiber start (m)');
plt.title("URI OCE Acoustic Tank Profile");
plt.savefig(outfname)
```
