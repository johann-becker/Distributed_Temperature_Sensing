## Fiber Homogeneity
#### This script takes in .csv inputs from the DTS system: probe Temperature measurements & Temperature measurements along the fiber to find the difference and percent differences along the fiber over time.
##### Inputs: Probe data .csv & Fiber measurements .csv
##### Outputs: diff -- The temperature difference; pdiff -- The percent difference; plots of diff and pdiff along fiber

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import datetime
```

```python
#User Input
pfname = 'Probe.csv' #Input Probe data file name .csv
ffname = 'Fiber.csv' #Input Fiber data file name .csv
tdif_out = 'temp_diff.png' #Output Temperature Difference name (include file type)
pdif_out = 'percent_diff.png' #Output Percent Difference name (include file type)
```

```python
#Load and format data
probe= pd.read_csv(pfname,header=0, dtype={0:'str'})
fiber= pd.read_csv(ffname,header=1)
probe['time'] = pd.to_datetime(probe['time'])
fiber.set_index(fiber['Distance'],inplace=True)
```

```python
# Calculate the temperature difference between the probe and fiber, and find the percent difference.
diff = fiber['Temperature.30']-probe['Probe'][30]
pdiff = (diff/probe['Probe'][30])*100
```

```python
#Plot the temperature differences along the fiber
plt.figure(figsize=(15,5))
plt.plot(fiber['Distance'],diff)
plt.xlabel('Distance along Fiber  (m)')
plt.ylabel('Temperature Difference $^\circ$C');
plt.title("Temperature Difference along Fiber");
plt.savefig(tdif_out)
```

```python
# Plot the % differences along the fiber
plt.figure(figsize=(15,5))
plt.plot(fiber['Distance'],pdiff)
plt.xlabel('Distance along Fiber (m)')
plt.ylabel('Percent Difference');
plt.title("Percent Difference along Fiber");
plt.savefig(pdif_out)
```
